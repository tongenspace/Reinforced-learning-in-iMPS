```python
import jax
import jax.numpy as jnp
from jax import random
import numpy as np

seed = 0
key = random.PRNGKey(seed)
Fs =  random.normal(key, (40,40))

a_jnp,b_jnp,c_jnp =  jnp.linalg.svd(Fs)
a_np, b_np, c_np = np.linalg.svd(Fs)

print("a:",jnp.linalg.norm(a_jnp-a_np))

print("b:",jnp.linalg.norm(b_jnp-b_np))

print("c:",jnp.linalg.norm(c_jnp-c_np))

print(b_jnp)
print(b_np)
```
