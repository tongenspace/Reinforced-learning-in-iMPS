```python
import jax
import jax.numpy as jnp
from jax import random
import numpy as np

seed = 0
key = random.PRNGKey(seed)
Fs =  random.normal(key, (40,40))

S_jnp,V_jnp,D_jnp = jnp.linalg.svd(Fs)

S_np, V_np, D_np  = np.linalg.svd(Fs)

print("Difference in S: ",jnp.linalg.norm(S_jnp-S_np))

print("Difference in V: ",jnp.linalg.norm(V_jnp-V_np))

print("Difference in D: ",jnp.linalg.norm(D_jnp-D_np))

print(V_jnp)
print(V_np)

print("Jax Version ",jax.__version__)
```
